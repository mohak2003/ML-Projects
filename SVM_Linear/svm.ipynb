```python
train = pd.read_csv("/kaggle/input/predicting-pulsar-starintermediate/pulsar_data_train.csv")
train

#importing dataset
```

```python
x_train, y_train = train.drop(['target_class'], axis=1), train['target_class']
x_train

#segregating features and target values
```

```python
x_train.isna().sum()

#dataset has several null spaces
```

```python
x_train.fillna(value=x_train.mean(), inplace=True)
x_train.isna().sum()

#filled nan values with column mean
```

```python
from sklearn.svm import SVC

model = SVC()  # C=1

model.fit(x_train,y_train)

test = pd.read_csv("/kaggle/input/predicting-pulsar-starintermediate/pulsar_data_train.csv")
x_test, y=test = test.drop(['target_class'],axis=1), test['target_class']

#fit the training dataset into the model
```

```python
x_test.isna().sum()

#nan values found in test dataset
```

```python
x_test.fillna(value = x_test.mean(), inplace = True)
x_test.isna().sum()

#filled nan spaces
```

```python
y_prediction = model.predict(x_test)

match = 0 
for i in range(len(y_test)):
               if (y_prediction[i]==y_test[i]):
                   match+=1
 
accuracy =  match/len(y_test)*100
accuracy

#printed accuracy of training model
```
