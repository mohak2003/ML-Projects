```python
#loading dataset
dataset = pd.read_csv("/kaggle/input/disney-movies-and-tv-shows/disney_plus_titles.csv")
dataset
```

```python
#dropping ineffective features
features = dataset.drop(['show_id','type','country','date_added','rating','duration','release_year'], axis =1)
features
```

```python
#checking null values
features.isnull().sum()
```

```python
#removing null values
features.dropna(inplace = True)
features.isna().sum()
```

```python
#resetting index
features =features.reset_index(drop = True)
features
```

```python
#creating combined-text feature to vectorize
features['metadata'] = features['title'] + ' ' + features['director'] + ' ' + features['cast'] + ' ' + features['listed_in'] + ' ' + features['description']
features['metadata']
```

```python
#importing vectorizer function for library to convert textual features to numerical values
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
```

```python
#applying vectorizer function and obtaining feature matrix
def create_feat_matrix(data, vectorizer):
    vocab = vectorizer.fit_transform(data)
    return pd.DataFrame(vocab.toarray(),columns = vectorizer.get_feature_names_out())
```

```python
feat_matrix = create_feat_matrix(features['metadata'], vectorizer)
feat_matrix
```

```python
#applying function to find all possible dot products of feature arrays of movie pairs
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(vocab)
similarity_matrix
```

```python
# For example finding movies similiar to 'X-Men Origins: Wolverine' (S.No.= 875)
sno = 875

# adding a counter to identify movie serial nos after sorting similarity values 
close_match = enumerate((similarity_matrix[sno].tolist()))
```

```python
#sorting the list by similarity values
close_match = sorted(close_match, key = lambda close_match : -close_match[1])

#viewing first 40 values
close_match[:40]
```

```python
#printing top 6 similiar movies

for i in range(6):
    index = close_match[i][0]
    print(features['title'][index])

```
