```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
src = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
dataset = np.genfromtxt(src, delimiter=",", dtype = str)

#  optimal value of k is sqrt(N)
k = int(np.sqrt(len(dataset))) + 1

print(dataset)
```

```python
#removed 'label' row to get features dataset
features = np.delete(dataset,-1,1) 

features = np.zeros([len(dataset),len(dataset[0])-1],dtype="float")

for i in range(len(features)):
    for j in range(len(features[0])):
        features[i][j] = float(dataset[i][j])


print(features[:20])
```

```python
#encoding labels to numeric values
for i in range(150):
    if dataset[i][4] == "Iris-setosa":
        dataset[i][4] = 0
    elif dataset[i][4] == 'Iris-versicolor':
        dataset[i][4] = 1
    elif dataset [i][4] == 'Iris-virginica':
        dataset[i][4] = 2

x_train, x_test = dataset[:110,:], dataset[110:,:]
print(x_train[:20])

```

```python
def dist(p1,p2):
    
    dist2 = 0
    
    for i in range(4):
        dist2 += (float(p1[i])-float(p2[i]))**2

    return(np.sqrt(dist2))
```

```python
#creating arrays of distance vectors
distArr = np.zeros([110,2])
for i in range(110):    
    distArr[i][1] = i

#x_test[5] as ex
for i in range(110):    
    distArr[i][0] = dist(x_train[i],x_test[5])

print(distArr[:20])
```

```python
sortarr = distArr[:,0].argsort()
print(sortarr)

Nearest_Neighbours = np.zeros([k,1],)
  
for i in range(k):
    
    Nearest_Neighbours[i] = dataset[sortarr[i]][4]

print(Nearest_Neighbours)
```

```python
prediction = max(Nearest_Neighbours)


print(f"The predicted flower type is {prediction}, The actual flower type was {x_test[5][4]}")
```
